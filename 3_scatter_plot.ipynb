Scatter , Swarm Plot , Box Plot , Count Plot , Violin Plot

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sales = pd.read_excel("D:\\Datasets\\sales\\sales.xlsx")
```

```python
sales.head()
```

```python
sns.catplot(x='Ship Mode' , y='Sales' , data=sales )
```

```python
sns.catplot(x='Ship Mode' , y='Sales' , data=sales ,jitter=False )#jitter is the magnitube width
```

```python
sns.catplot(x='Ship Mode' , y='Sales' , data=sales.query("Profit > 1000"), jitter=False)
```

```python
sns.catplot(x='Ship Mode' , y='Sales' , data=sales.query("Profit>1000") , kind='swarm')
```
