Facet-Working with multiple graphs at a time

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sales = pd.read_excel("D:\\Datasets\\sales\\sales.xlsx")
```

```python
sales.head()
```

```python
sns.relplot(x = 'Sales', y='Profit' , data=sales)
```

```python
sns.relplot(x = 'Sales', y='Profit' , data=sales , col='Customer Segment',col_wrap=2)
```

```python
sns.relplot(x = 'Sales', y='Profit' , data=sales , col='Customer Segment',hue='Ship Mode',col_wrap=2,kind='line')
```

```python
sns.relplot(x = 'Sales', y='Profit' , data=sales , col='Customer Segment',row='Product Category',hue='Ship Mode')
```
