Bar PLot and Point plot

```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel("D:\\Datasets\\sales\\sales.xlsx")
```

```python
sales.head()
```

```python
sales.columns
```

```python
sns.catplot(x='Product Category',y='Sales',data=sales,kind='bar')
```

```python
sns.catplot(x='Product Category',y='Sales',hue='Ship Mode',data=sales,kind='bar')
```

```python
sns.catplot(x='Product Category',y='Sales',hue='Ship Mode',col='Customer Segment',data=sales,kind='bar')
```

```python
sns.catplot(x='Region',y='Sales',data=sales,hue='Product Category',kind='bar',aspect=3,errorbar=None)
```
