Box Chart

```python
import pandas as pd
import seaborn as sns
```

```python
sales = pd.read_excel("D:\\Datasets\\sales\\sales.xlsx")
```

```python
sales.head()
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data=sales,kind='box')
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data=sales,hue='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data=sales,hue='Ship Mode',hue_order=["Delivery Truck","Regular Air","Express Air"],kind='box')
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data=sales,row='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category' , y='Shipping Cost' , data=sales,col='Ship Mode',row='Customer Segment',kind='box')
```
