```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sales = pd.read_excel("D:\\Datasets\\sales\\sales.xlsx")
```

```python
sales.head()
```

```python
sns.relplot(x='Sales',y='Profit',data=sales)
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,hue='Order Priority')
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,hue='Order Priority',style='Order Priority')
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,hue='Order Priority',style='Ship Mode')
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,size='Discount')
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,size='Discount',sizes=(25,200))
```

```python
sns.relplot(x='Order Date',y='Sales',data=sales,kind='line')
```

```python
sales['year'] = sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line')
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',errorbar=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',errorbar='sd')
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',estimator=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',estimator=sum)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',errorbar=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Product Category',errorbar=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Region',errorbar=None)
```

```python
sns.relplot(x='year',y='Sales',data=sales,kind='line',hue='Ship Mode',col='Region',errorbar=None,col_wrap=4)
```
